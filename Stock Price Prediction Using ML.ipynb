```python
splitted = df['Date'].str.split('/', expand=True)

df['day'] = splitted[1].astype('int')
df['month'] = splitted[0].astype('int')
df['year'] = splitted[2].astype('int')

df.head()
```

```python
df['is_quarter_end'] = np.where(df['month']%3==0,1,0)
df.head()
```

```python
data_grouped = df.drop('Date', axis=1).groupby('year').mean()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
sns.barplot(data = data_grouped['Open'].reset_index(), x = 'year', y = 'Open')

plt.subplot(2,2,2)
sns.barplot(data = data_grouped['High'].reset_index(), x = 'year', y = 'High')

plt.subplot(2,2,3)
sns.barplot(data = data_grouped['Low'].reset_index(), x = 'year', y = 'Low')

plt.subplot(2,2,4)
sns.barplot(data = data_grouped['Close'].reset_index(), x = 'year', y = 'Close')

plt.tight_layout()
plt.show()
```

```python
df.drop('Date', axis=1).groupby('is_quarter_end').mean()
```

```python
df['open-close'] = df['Open'] - df['Close']
df['low-high'] = df['Low'] - df['High']
df['target'] = np.where(df['Close'].shift(-1) > df['Close'], 1, 0)
```

```python
plt.pie(df['target'].value_counts().values, labels=[0, 1], autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10, 10))

# As our concern is with the highly correlated features only so, we will visualize our heatmap as per that criteria only.
sns.heatmap(df.drop('Date', axis=1).corr() > 0.9, annot=True, cbar=False)
plt.show()
```

```python
features = df[['open-close', 'low-high', 'is_quarter_end']]
target = df['target']

scaler = StandardScaler()
features = scaler.fit_transform(features)

X_train, X_valid, Y_train, Y_valid = train_test_split(
	features, target, test_size=0.1, random_state=2022)
print(X_train.shape, X_valid.shape)
```

```python
models = [LogisticRegression(), SVC(
kernel='poly', probability=True), XGBClassifier()]

for i in range(3):
  models[i].fit(X_train, Y_train)

  print(f'{models[i]} : ')
  print('Training Accuracy : ', metrics.roc_auc_score(
    Y_train, models[i].predict_proba(X_train)[:,1]))
  print('Validation Accuracy : ', metrics.roc_auc_score(
    Y_valid, models[i].predict_proba(X_valid)[:,1]))
  print()
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(models[0], X_valid, Y_valid)
plt.show()
```
